```python
def remove_cosmic_points(dataframe, frame_number, threshold_factor=0.07):

    dataframe = dataframe.copy()
    #dataframe = dataframe[::-1].reset_index(drop=True)
    
    rows_to_drop = []
    
    for i in range(1, len(dataframe)):
        if abs(dataframe[f"Frame{frame_number} Intensity"].iloc[i] - dataframe[f"Frame{frame_number} Intensity"].iloc[i - 1]) > 125: 
            rows_to_drop.append(i)
            
    dataframe = dataframe.drop(rows_to_drop)
    
    #dataframe = dataframe[dataframe[f"Frame{frame_number} Intensity"] >= 625]
            
    return dataframe
```

```python
def gaussian(x, A, mu, sigma):
    return A * np.exp(-(x - mu)**2 / (2 * sigma**2))

def lorentzian(x, A, mu, gamma):
    return A / (1 + ((x - mu) / gamma)**2)
```

```python
from scipy.optimize import curve_fit

df_test_GL = df.copy()
df_test_GL = df_test_GL[['Wavenumber', 'Frame1 Intensity']]

p0_gaussian = [np.max(df_test_GL['Frame1 Intensity']), np.mean(df_test_GL['Wavenumber']), 1]
p0_lorentzian = [np.max(df_test_GL['Frame1 Intensity']), np.mean(df_test_GL['Wavenumber']), 1]

bounds = ([0, min(df_test_GL['Wavenumber']), 0], [np.inf, max(df_test_GL['Wavenumber']), np.inf])

params_gaussian, _ = curve_fit(gaussian, df_test_GL['Wavenumber'], df_test_GL['Frame1 Intensity'], p0=p0_gaussian, bounds=bounds, maxfev=800)

params_lorentzian, _ = curve_fit(lorentzian, df_test_GL['Wavenumber'], df_test_GL['Frame1 Intensity'], p0=p0_lorentzian, bounds=bounds, maxfev=800)

fitted_baseline_gaussian = gaussian(df_test_GL['Wavenumber'], *params_gaussian)
fitted_baseline_lorentzian = lorentzian(df_test_GL['Wavenumber'], *params_lorentzian)

df_test_GL.loc[:, 'Corrected_Intensity_Gaussian'] = df_test_GL['Frame1 Intensity'] - fitted_baseline_gaussian
df_test_GL.loc[:, 'Corrected_Intensity_Lorentzian'] = df_test_GL['Frame1 Intensity'] - fitted_baseline_lorentzian

fig, axes = plt.subplots(1, 2, figsize=(25, 15))

axes[0].plot(df_test_GL['Wavenumber'], df_test_GL['Corrected_Intensity_Gaussian'], color='blue', label='Corrected Spectrum (Gaussian)')
axes[0].set_xlabel('Wavenumber')
axes[0].set_ylabel('Corrected Intensity')
axes[0].legend()
axes[0].axhline(0, color='black', linestyle='--')
axes[0].set_title('Gaussian Fitting')

axes[1].plot(df_test_GL['Wavenumber'], df_test_GL['Corrected_Intensity_Lorentzian'], color='green', label='Corrected Spectrum (Lorentzian)')
axes[1].set_xlabel('Wavenumber')
axes[1].set_ylabel('Corrected Intensity')
axes[1].legend()
axes[1].axhline(0, color='black', linestyle='--')
axes[1].set_title('Lorentzian Fitting')

plt.tight_layout()
plt.show()
```

```python
from scipy.special import wofz

def voigt(x, A, mu, sigma, gamma):
    z = ((x - mu) + 1j * gamma) / (sigma * np.sqrt(2))  
    return A * np.real(wofz(z))

def gl_mixed(x, A, mu, sigma, gamma, alpha):
    gaussian = A * np.exp(-(x - mu)**2 / (2 * sigma**2))
    lorentzian = A / (1 + ((x - mu) / gamma)**2)
    return (1 - alpha) * gaussian + alpha * lorentzian
```

```python
df_test_VGL = df.copy()
df_test_VGL = df_test_VGL[['Wavenumber', 'Frame1 Intensity']]

p0_voigt = [np.max(df_test_VGL['Frame1 Intensity']), np.mean(df_test_VGL['Wavenumber']), 1, 1]
p0_gl_mixed = [np.max(df_test_VGL['Frame1 Intensity']), np.mean(df_test_VGL['Wavenumber']), 1, 1, 0.5]

bounds_voigt = ([0, min(df_test_VGL['Wavenumber']), 0, 0], [np.inf, max(df_test_VGL['Wavenumber']), np.inf, np.inf])
bounds_gl_mixed = ([0, min(df_test_VGL['Wavenumber']), 0, 0, 0], [np.inf, max(df_test_VGL['Wavenumber']), np.inf, np.inf, 1])

params_voigt, _ = curve_fit(voigt, df_test_VGL['Wavenumber'], df_test_VGL['Frame1 Intensity'], p0=p0_voigt, bounds=bounds_voigt, maxfev=10000)

params_gl_mixed, _ = curve_fit(gl_mixed, df_test_VGL['Wavenumber'], df_test_VGL['Frame1 Intensity'], p0=p0_gl_mixed, bounds=bounds_gl_mixed, maxfev=10000)

fitted_baseline_voigt = voigt(df_test_VGL['Wavenumber'], *params_voigt)
fitted_baseline_gl_mixed = gl_mixed(df_test_VGL['Wavenumber'], *params_gl_mixed)

df_test_VGL.loc[:, 'Corrected_Intensity_Voigt'] = df_test_VGL['Frame1 Intensity'] - fitted_baseline_voigt
df_test_VGL.loc[:, 'Corrected_Intensity_GL_Mixed'] = df_test_VGL['Frame1 Intensity'] - fitted_baseline_gl_mixed

fig, axes = plt.subplots(1, 2, figsize=(25, 15))

axes[0].plot(df_test_VGL['Wavenumber'], df_test_VGL['Corrected_Intensity_Voigt'], color='blue', label='Corrected Spectrum (Voigt)')
axes[0].set_xlabel('Wavenumber')
axes[0].set_ylabel('Corrected Intensity')
axes[0].legend()
axes[0].axhline(0, color='black', linestyle='--')
axes[0].set_title('Voigt Fitting')

axes[1].plot(df_test_VGL['Wavenumber'], df_test_VGL['Corrected_Intensity_GL_Mixed'], color='green', label='Corrected Spectrum (G-L Mixed)')
axes[1].set_xlabel('Wavenumber')
axes[1].set_ylabel('Corrected Intensity')
axes[1].legend()
axes[1].axhline(0, color='black', linestyle='--')
axes[1].set_title('G-L Mixed Fitting')

plt.tight_layout()
plt.show()
```

```python
from scipy.integrate import trapz

def calculate_fwhm(wavenumbers, intensities):
    half_max = np.max(intensities) / 2

    above_half_max = np.where(intensities >= half_max)[0]

    if len(above_half_max) > 1:
        fwhm = wavenumbers.iloc[above_half_max[-1]] - wavenumbers.iloc[above_half_max[0]]
    else:
        fwhm = 0 
    return fwhm

def calculate_r2_linear(x, y, model_func, params):
    y_pred = model_func(x, *params)
    ss_total = np.sum((y - np.mean(y))**2)
    ss_residual = np.sum((y - y_pred)**2)
    return 1 - (ss_residual / ss_total)

def calculate_r2_non_linear(x, y, model, params):
    y_pred = model(x, *params)
    ss_res = np.sum((y - y_pred) ** 2)
    ss_tot = np.sum((y - np.mean(y)) ** 2)
    return 1 - (ss_res / ss_tot)

def calculate_rmse(observed, fitted):
    residuals = observed - fitted
    return np.sqrt(np.mean(residuals ** 2))
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
plt.figure(figsize=(25, 15))

for i in range(num_frames):
    df_subset1 = df[['Wavenumber', f'Frame{i+1} Intensity']]
    df_subset1 = df_subset1[df_subset1['Wavenumber'] >= 0]
    #df_subset1 = remove_cosmic_points(df_subset1, frame_number = i+1)
    #df_subset1 = remove_cosmic_points(df_subset1, frame_number = i+1)

    poly = PolynomialFeatures(degree=1)
    X_poly = poly.fit_transform(df_subset1['Wavenumber'].values.reshape(-1, 1))
    model = LinearRegression()
    model.fit(X_poly, df_subset1[f'Frame{i+1} Intensity'])
    fitted_baseline = model.predict(X_poly)

    rmse_linear = calculate_rmse(df_subset1[f'Frame{i+1} Intensity'], fitted_baseline)
    print(f"Frame {i+1} Linear RMSE: {rmse_linear:.4f}")

    r2_linear = calculate_r2_linear(df_subset1['Wavenumber'], df_subset1[f'Frame{i+1} Intensity'], lambda x, *params: model.predict(poly.transform(np.array(x).reshape(-1, 1))), [model.intercept_, *model.coef_])
    print(f"Frame {i+1} Linear R²: {r2_linear:.4f}")
    
    if r2_linear < 0.75:
        p0_voigt = [np.max(df_subset1[f'Frame{i+1} Intensity']), np.mean(df_subset1['Wavenumber']), 1, 1]
        bounds_voigt = ([0, min(df_subset1['Wavenumber']), 0, 0], [np.inf, max(df_subset1['Wavenumber']), np.inf, np.inf])
        params_voigt, _ = curve_fit(voigt, df_subset1['Wavenumber'], df_subset1[f'Frame{i+1} Intensity'], p0=p0_voigt, bounds=bounds_voigt, maxfev=800)
        
        fitted_baseline = voigt(df_subset1['Wavenumber'], *params_voigt)
        
        label = 'Voigt Fit'
        
        rmse_voigt = calculate_rmse(df_subset1[f'Frame{i+1} Intensity'], fitted_baseline)
        print(f"Frame {i+1} Voigt RMSE: {rmse_voigt:.4f}")
        
        r2_voigt = calculate_r2_non_linear(df_subset1['Wavenumber'], df_subset1[f'Frame{i+1} Intensity'], gaussian, params_gaussian)
        print(f"Frame {i+1} Voigt R²: {r2_voigt:.4f}")

    else: 
        label = 'Linear Fit'
        
    df_subset1.loc[:, f'Frame{i+1} Corrected Intensity'] = df_subset1[f'Frame{i+1} Intensity'] - fitted_baseline
    #df_subset1 = df_subset1[(df_subset1["Wavenumber"] >= 1330) & (df_subset1["Wavenumber"] <= 1400)]
    
    plt.scatter(df_subset1['Wavenumber'], df_subset1[f'Frame{i+1} Corrected Intensity'], label=f'Frame {i+1} ({label})')


plt.title('Combined Scatter Plot')
plt.xlabel('Wavenumber')
plt.ylabel(f'Frame{i+1} Intensity')

# Legend to differentiate frames
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

```python
import dash
from dash import dcc, html
import plotly.graph_objects as go
import numpy as np
import pandas as pd

grid_size = 19

intensity_matrix = np.zeros((grid_size, grid_size))
wavenumber_matrix = np.zeros((grid_size, grid_size))

for i in range(num_frames):
    df_range = df[['Wavenumber', f'Frame{i+1} Intensity']]
    df_range = df_range[df_range['Wavenumber'] >= 0]
    #df_range = remove_cosmic_points(df_range, frame_number = i+1)

    max_intensity = df_range[f'Frame{i+1} Intensity'].max()
    max_wavenumber = df_range.loc[df_range[f'Frame{i+1} Intensity'].idxmax(), 'Wavenumber']

    row = (i) // grid_size
    col = (i) % grid_size
    row = grid_size - row - 1
    intensity_matrix[row, col] = max_intensity
    wavenumber_matrix[row, col] = max_wavenumber

app = dash.Dash(__name__)

def plot_frame(frame_index):
    row = grid_size - (frame_index // grid_size) - 1 
    col = frame_index % grid_size
    
    df_range = df[['Wavenumber', f'Frame{i+1} Intensity']]
    df_range = df_range[df_range['Wavenumber'] >= 0]
    #df_range = remove_cosmic_points(df_range, frame_number = i+1)
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=df_range['Wavenumber'],
        y=df_range[f'Frame{i+1} Intensity'],
        mode='markers',
        marker=dict(color='blue'),
        name=f'Frame {frame_index + 1}'
    ))
    fig.update_layout(
        title=f'Frame {frame_index + 1} - Intensity vs Wavenumber',
        xaxis_title='Wavenumber',
        yaxis_title=f'Frame{i+1} Intensity'
    )
    return fig

app.layout = html.Div([ 
    html.H1("Interactive Intensity Heatmap and Frame Viewer"),
    
    dcc.Graph(
        id='heatmap-intensity',
        figure=go.Figure(
            data=[go.Heatmap(
                z=intensity_matrix,
                colorscale='Reds',
                name=f'Frame{i+1} Intensity',
                hovertemplate='Row: %{y}<br>Column: %{x}<br>Intensity: %{z}<br>Wavenumber: %{customdata}<br><extra></extra>',
                customdata=wavenumber_matrix,
            )],
            layout=go.Layout(
                title='Heatmap of Maximum Intensity',
                xaxis_title='Column Index',
                yaxis_title='Row Index',
            )
        )
    ),
    
    dcc.Graph(
        id='frame-plot',
    ),
])

@app.callback(
    dash.dependencies.Output('frame-plot', 'figure'),
    [dash.dependencies.Input('heatmap-intensity', 'clickData')]
)
def update_graph(clickData):
    if clickData:
        point = clickData['points'][0]
        row = grid_size - point['y'] - 1
        col = point['x']
        frame_index = row * grid_size + col
        return plot_frame(frame_index)
    return {}

if __name__ == '__main__':
    app.run_server(debug=True, port=8051)
```
